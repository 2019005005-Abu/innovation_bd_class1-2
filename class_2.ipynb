```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt; 
import seaborn as sns;
import cv2;

img_path='../dataset/Flower Dataset/flowers/test/1/image_06743.jpg';
img=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE);
threshold_binary=cv2.threshold(img,127,255,cv2.THRESH_BINARY);
threshold_binary_inv=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV);##Binary Inverse
threshold_trunc=cv2.threshold(img,127,255,cv2.THRESH_TRUNC);#Binary Truncate
#cv2.THRESH_BINARY_INV+cv2.THRESH_BINARY+cv2.THRESH_TRUNC+cv2.THRESH_TOZERO+cv2.THRESH_TOZERO_INV
threshold_thesh_terzero=cv2.threshold(img,127,255,cv2.THRESH_TOZERO);#Binary Tozero
threshold_terzoro_inv=cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV);

ret,thesh1=threshold_binary;
ret,thesh2=threshold_binary_inv;
ret,thesh3=threshold_trunc;
ret,thesh4=threshold_thesh_terzero;
rest,thresh5=threshold_terzoro_inv;



titile=[
    'Original Image',
    'Binary',
    'Binary Inverse',
    'Truncate',
    'Tozero',
    'Tozero Inverse'
]

images=[
    img,
    thesh1,
    thesh2,
    thesh3,
    thesh4,
    thresh5
]

for i in range(6):
    plt.subplot(2,3,i+1);
    plt.imshow(images[i],cmap='gray');
    plt.title(titile[i]);
    plt.xticks([])
    plt.yticks([])
plt.show();


```
